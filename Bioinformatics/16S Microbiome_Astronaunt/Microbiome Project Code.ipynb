##Quality control using CutAdapt V4.2 and DADA2

Cutadapt plugin in qiime2 was used to trim the set of primers from the paired end data.

```python
# Load qiime2 environment in the same directory as the data and open the jupyter notebook.
#Importing the demultiplexed paired end reads into qiime2.
!qiime tools import \
  --type "SampleData[PairedEndSequencesWithQuality]" \
  --input-format PairedEndFastqManifestPhred33V2 \
  --input-path Sequencing_data/manifest_tmp.tsv \
  --output-path demux_seqs.qza
```

```python
#Importing the cutadapt plugin in qiime2.
from qiime2.plugins import cutadapt
```

```python
#Using qiime2 plugin cutadapt to remove the primers.
!qiime cutadapt trim-paired \
  --i-demultiplexed-sequences demultiplexed-seqs.qza \
  --p-front-f CCTACGGGNGGCWGCAG \
  --p-front-r GACTACHVGGGTATCTAATCC \
  --p-error-rate 0 \
  --o-trimmed-sequences trimmed-seqs.qza \
  --verbose
```

DADA2 plugin for qiime2 was used for quality control. DADA2 helps with  filtering out noisy sequences, correcting errors in marginal sequences, removing chimeric sequences, removing singletons, joining denoised paired-end reads, and then dereplicating those sequences (https://docs.qiime2.org/2023.2/tutorials/overview/).

The parameter used below for the forward and reverse primers were chosen based on the paper from which the data was taken.

```python
#DADA2 was used to conduct quality control.
!qiime dada2 denoise-paired \
  --i-demultiplexed-seqs trimmed-seqs.qza \
  --p-trim-left-f 17 \
  --p-trim-left-r 21 \
  --p-trunc-len-f 280 \
  --p-trunc-len-r 220 \
  --p-trunc-q 2 \
  --p-chimera-method  "consensus" \
  --p-max-ee-f 2 \
  --p-max-ee-r 2 \
  --o-table space_table.qza \
  --o-representative-sequences space_paired.qza \
  --o-denoising-stats space_denoising_stats.qza
```

##Taxonomy Classification: Training the Classifier from Greengenes2 using Naive Bayes in Qiime2

First, the entire backbone sequence was obtained from Greengenes2 release.

```python
!wget http://ftp.microbio.me/greengenes_release/2022.10/2022.10.backbone.full-length.fna.qza
```

p-trunc-len was removed based on the suggestion by the tutorial about unequal lengths for the paired end sequences.
The lengths for 400 and 500 were chosen as per the forum chat and suggestion of a relaxed interval or allowed amplicon size to not miss biologically relevant data:https://forum.qiime2.org/t/picking-values-for-p-min-length-and-p-max-length-in-qiime-feature-classifier-extract-reads/20912.

```python
#Extracting reads specific to the v3-v4 region.
!qiime feature-classifier extract-reads \
  --i-sequences 2022.10.backbone.full-length.fna.qza \
  --p-f-primer CCTACGGGNGGCWGCAG \
  --p-r-primer GACTACHVGGGTATCTAATCC \
  --p-min-length 400 \
  --p-max-length 500 \
  --o-reads gg2-ref-seqs.qza
```

```python
#Classifier was trained using Naiye Bayes
!qiime feature-classifier fit-classifier-naive-bayes \
  --i-reference-reads gg2-ref-seqs.qza \
  --i-reference-taxonomy 2022.10.backbone.tax.qza \
  --o-classifier classifier.qza
```

##*De Novo* Phylogenetic Tree using IQTree

To create a *De Novo* Phylogenetic tree, multiple sequence alignment of the ASVs is done using MAFFT and IQTree is used to create the most probable phylogenetic tree from that alignment. IQTree is based on maximum likelihood which means that it will find the tree most likely to be possible given the alignment from multiple possible trees.

```python
!qiime phylogeny align-to-tree-mafft-iqtree \
  --i-sequences space_paired.qza \
  --o-alignment space_aligned.qza \
  --o-masked-alignment space_masked_aligned.qza \
  --o-tree space_unrooted_tree.qza \
  --o-rooted-tree space_rooted_tree.qza \
  --p-alrt 1000 \
  --p-seed 42 \
  --p-n-threads 8
```

##Alpha Diversity: Shannon

First the feature table is visualized to look at the minimum feature count and get a general distribution.

```python
!qiime feature-table summarize \
  --i-table space_table.qza \
  --o-visualization space_table.qzv \
  --m-sample-metadata-file space_tmp.tsv
```

Alpha rarefaction curves are generated at multiple sequencing depths and gives a visual of how the Shannon entropy changes at different depths for the three flight groups ad the control. All the curves plateau at ~500.

```python
#Creating the rarefaction curves with a maximum depth of 4700.
!qiime diversity alpha-rarefaction \
  --i-table space_table.qza \
  --i-phylogeny space_rooted_tree.qza \
  --p-max-depth 4700 \
  --m-metadata-file space_tmp.tsv \
  --o-visualization space_shannon_alpha_rarefaction.qzv
```

Looking at the feature table and the curves, a sampling depth of 4700 was chosen which excludes the controls from analyses and all other feature counts are above this count. Also, the curves are a plateau at this point. Rarefying at this depth would make all the feature counts 4700. 

```python
#Rarefying.
!qiime feature-table rarefy \
  --i-table space_table.qza \
  --p-sampling-depth 4700 \
  --o-rarefied-table space_rarefied_table.qza
```

Shannon metric is used for alpha diversity analysis which uses both evenness and richness of the community to give a measure of diversity within sample.

```python
#Plugin for alpha diversity.
!qiime diversity alpha \
    --i-table space_rarefied_table.qza \
    --p-metric shannon \
    --o-alpha-diversity space_shannon_vector.qza
```

```python
#Exporting and creating a tab separated file.
!qiime tools export --input-path space_shannon_vector.qza --output-path .
!mv alpha-diversity.tsv space_shannon.tsv
```

Alpha diversity for each of the groups was calculated by grouping based on flight time. A Kruskal Wallis test is conduted to see if the different groups have different alpha diversity. A p value and H statistic is computed and pairwise comparisons for each of the groups are done with post hoc tests to look for difference between a pair of groups.

```python
!qiime diversity alpha-group-significance \
    --i-alpha-diversity space_shannon_vector.qza \
    --m-metadata-file space_tmp.tsv \
    --o-visualization space_shannon_group_significance.qzv
```

```python
#Exporting and creating a tab separated file.
!qiime tools export --input-path space_shannon_vector.qza --output-path .
!mv alpha-diversity.tsv space_shannon.tsv
```

##Beta Diversity using Weighted-UniFrac

To measure the difference in diversity between samples, beta diversity was measured. The metric used was Weighted-UniFrac that is phylogeny based and the branches of the phylogeny are abundance weighted.

```python
#beta-phylogenetic to tell qiime2 to use the phylogeny based metrics. A distance matrix is created at the end.
!qiime diversity beta-phylogenetic \
    --i-table space_rarefied_table.qza \
    --i-phylogeny space_rooted_tree.qza \
    --p-metric weighted_unifrac \
    --o-distance-matrix space_weighted_unifrac_distance_matrix.qza
```

```python
#Exporting the matrix and creating a tab separated file for the same.
!qiime tools export --input-path space_weighted_unifrac_distance_matrix.qza --output-path .
!mv distance-matrix.tsv space_weighted_unifrac.tsv
```

```python
#Visualizing the tsv file created.
!head space_weighted_unifrac.tsv
```

To look for differences based on flight time, Adonis test was conducted. Adonis test in qiime2 is wrapper of the adonis function in the vegan R package and therefore, cannot accept files with '-'. Therefore, files are first changed to the desirable format.

```python
#Removing '-' from the metadata file.
!cat space_tmp.tsv | tr -d '-' > space_tmp_rev.tsv
```

```python
#Running the Adonis test using qiime2 plugin.
!qiime diversity adonis \
  --i-distance-matrix space_weighted_unifrac_distance_matrix.qza \
  --m-metadata-file space_tmp_rev.tsv \
  --p-formula "FactorValueTime" \
  --o-visualization weighted_unifrac_Time_adonis.qzv
```

The Adonis test provides the F statistic which is the ratio of intergroup to intragroup variance and, in this case, is more than 1 but not large enough. The R2 value is the proportion of variance between the beta diversity of samples explained by the grouping variable in our case flight time which is ~5% and not very large.

##Differential Abundance Analysis using ANCOM-BC

The taxa table was first collapsed to species, genus and family level (l7,l6 and l5 respectively).

```python
#Collapsing the non-rarefied table to the specific taxa levels using a bash cell and for loop.
%%bash
for i in {5,6,7}; do
  qiime taxa collapse \
    --i-table space_table.qza \
    --i-taxonomy space_taxonomy.qza \
    --p-level $i \
    --o-collapsed-table l$i.qza
done
```

```python
#Creating a visualization for each of the .qza files generated in the last cell.
%%bash
for i in {5,6,7}; do
  qiime feature-table summarize \
    --i-table l$i.qza \
    --o-visualization l$i.qzv \
    --m-sample-metadata-file space_tmp.tsv
done
    
```

```python
#Changing the way negative controls are labelled in the table.
!sed -i 's/notapplicable/control/g' space_tmp_rev.tsv
```

ANCOM-BC takes into consideration the compositionality of the microbiome and how changes in the sampling fraction used can change results of the abundance analysis. ANCOM-BC can only compare two groups at a time and therefore, a reference is chosen to be able to make comparisons between the three flight times. Since one of the questions of the study is to look whether during flight, the anxiety and depression levels, or such mental health issues are aggravated or more likely to occur due to space flight as a stressor, during flight is chosen as the reference for the ANCOM-BC comparison. ANCOM-BC will provide the log fold changes, p values and the q values. 

```python
#Running ANCOM-BC on family level given the previous largely insignificant results for genus level with the diversity metrics.
!qiime composition ancombc \
  --i-table l5.qza \
  --m-metadata-file space_tmp_rev.tsv \
  --p-reference-levels FactorValueTime::duringflight \
  --p-formula FactorValueTime \
  --o-differentials ancombc_space_time_during_flight_l5.qza
```

```python
#Creating a visualization of the ANCOM-BC results from the previous cell.
!qiime composition tabulate \
  --i-data ancombc_space_time_during_flight_l5.qza \
  --o-visualization ancombc_space_time_during_flight_l5.qzv
```

ANCOM-BC results can be visualized using another qiime2 plugin. To create this plot, the latest version of qiime2 (2023 version) was first installed. The visualization shows the taxa, at the family level, that are depleted or enriched with respect to during flight for both pre-flight and post-flight groups.

```python
#Creating the barplot.
!qiime composition da-barplot \
  --i-data ancombc_space_time_during_flight_l5.qza \
  --o-visualization ancombc_space_time_during_l5_barplot.qzv
```
